```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# configuración estética
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("sales.csv")
df.head()# Información general del DataFrame
```

```python
# ver información general
df.info()

# eliminar duplicados
df = df.drop_duplicates()

# valores nulos
df.isnull().sum()

# convertir fecha (si existe columna tipo "Date")
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
Análisis EDA (Exploratory Data Analysis) de un conjunto de datos de ventas utilizando Python y pandas. El código carga un archivo CSV llamado "sales.csv" en un DataFrame y muestra las primeras filas del DataFrame para obtener una visión general de los datos.
```

```python
# Ventas por mes
df['order date'] = pd.to_datetime(df['order date'])
df['month'] = df['order date'].dt.to_period('M')

monthly_sales = df.groupby('month')['sales'].sum()

plt.figure(figsize=(12,6))
monthly_sales.plot(linewidth=3)

plt.title("Monthly Sales Trend", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)

plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.savefig("../images/monthly_sales.png")
plt.show()
```

```python
#Top productos vendidos

top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Products by Sales")
plt.show()
```

```python
# Ventas por region

top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Products by Sales")
plt.show()
```

```python
# Mejores Clientes

top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_customers.plot(kind='bar')
plt.title("Top Customers")
plt.show()
```

## Conclusiones clave

- Las ventas muestran tendencias mensuales claras, lo que sugiere estacionalidad.

- Un número reducido de productos genera una gran parte de los ingresos totales.

- Algunas regiones superan significativamente a otras.

- Los clientes de alto valor contribuyen de forma desproporcionada a los ingresos.
